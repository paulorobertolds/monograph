```python
#importing required packages
import pandas as pd
import numpy as np

#importing the data
df = pd.read_csv('https://raw.githubusercontent.com/paulorobertolds/monograph/main/zaninData.csv' )


print(df.shape)
df.head()
```

```python
df=df.drop(df.columns[[17,18,19,20,21]],axis =1)
df.head()

```

```python
np.unique(df[df.columns[[6]]])
```

```python
np.unique(df[df.columns[[7]]])
```

```python
# Nomes das colunas
print(df.columns.values)
```

```python
# Remover colunas desnecessárias:
df=df.drop(['Data de extração','Base de dados','SETOR IBGE','Divisão','Divisão (Nome)','Grupo','Grupo (Nome)','Classe','Classe (Nome)','Subclasse','Denominação'],axis=1)
df.head()
```

```python
#df.rename(columns = {'Seção':'Setor', 'Seção (Nome)':'Nome do Setor'}, inplace = True)
```

```python
empresas_Cuiaba = df.loc[(df['Local']=='Cuiabá')&(df['Tipo'] == 'N° de empresas')]
empresas_Cuiaba.head()
```

```python
#Agrupar por seção
Empresas_Cuiaba=empresas_Cuiaba.groupby(['Ano','Seção','Seção (Nome)'],as_index = False)['Total'].sum().fillna(0)
Empresas_Cuiaba.head()
```

```python
funcionarios_Cuiaba = df.loc[(df['Local']=='Cuiabá')&(df['Tipo'] == 'N° de funcionários')]
funcionarios_Cuiaba.head()
```

```python
#Agrupar por seção
Funcionarios_Cuiaba=funcionarios_Cuiaba.groupby(['Ano','Seção','Seção (Nome)'],as_index = False)['Total'].sum().fillna(0)
Funcionarios_Cuiaba.head()
```
